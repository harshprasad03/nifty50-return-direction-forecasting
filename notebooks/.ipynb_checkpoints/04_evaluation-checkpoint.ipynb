```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

eval_df = pd.read_csv("../data/evaluation_data.csv")
```

```python
strategy_returns = eval_df["Actual_Return"] * eval_df["Prediction"]

strategy_cum = (1 + strategy_returns).cumprod()
buy_hold_cum = (1 + eval_df["Actual_Return"]).cumprod()
```

```python
plt.figure(figsize=(10,6))
plt.plot(strategy_cum, label="Strategy")
plt.plot(buy_hold_cum, label="Buy & Hold")
plt.legend()
plt.title("Strategy vs Buy & Hold")
plt.show()
```

```python
# Sharpe Ratio (assuming risk-free rate ~ 0)
sharpe = strategy_returns.mean() / strategy_returns.std()
print("Strategy Sharpe Ratio:", sharpe)

# Volatility
print("Strategy Volatility:", strategy_returns.std())
print("Buy & Hold Volatility:", eval_df["Actual_Return"].std())

# Maximum Drawdown
rolling_max = strategy_cum.cummax()
drawdown = (strategy_cum - rolling_max) / rolling_max
print("Maximum Drawdown:", drawdown.min())
```

### Transaction Costs

The strategy assumes zero transaction costs.

In reality, daily trading incurs brokerage, slippage,
and bid-ask spread costs.

Incorporating transaction costs may reduce net profitability.

## Final Financial Conclusion

The predictive model provides limited directional edge.

Strategy performance is comparable to Buy & Hold
and does not consistently outperform after risk adjustment.

These results are consistent with weak-form market efficiency,
where short-term index movements are difficult to predict.
