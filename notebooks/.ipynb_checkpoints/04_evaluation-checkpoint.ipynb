# 04 — Evaluation & Visualization
## Nifty 50 Return Direction Forecasting
This notebook loads saved models and generates evaluation plots
including accuracy comparison, ROC curves, confusion matrix,
and feature importance.

**Output:** outputs/figures/ and outputs/results/

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import os

# Load data
df = pd.read_csv('../data/nifty50_features.csv', index_col=0)
df.index = pd.to_datetime(df.index)

features = ['Return', 'Return_lag1', 'Return_lag2', 'Return_lag3',
            'Volatility_10', 'RSI', 'MA_Cross', 
            'MA5_20_ratio', 'MA20_50_ratio']

X = df[features]
y = df['Target']

# Time based split
split = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# Load models and scaler
scaler = joblib.load('../outputs/models/scaler.pkl')
lr = joblib.load('../outputs/models/logistic_regression.pkl')
rf = joblib.load('../outputs/models/random_forest.pkl')
xgb = joblib.load('../outputs/models/xgboost.pkl')

# Scale test data
X_test_scaled = scaler.transform(X_test)

print("Everything loaded successfully!")
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
import matplotlib.pyplot as plt
import os

os.makedirs('../outputs/figures', exist_ok=True)

# Get predictions
lr_pred = lr.predict(X_test_scaled)
rf_pred = rf.predict(X_test)
xgb_pred = xgb.predict(X_test)
baseline_pred = np.ones(len(y_test))

# Accuracies
models = ['Naive Baseline', 'Logistic Regression', 'Random Forest', 'XGBoost']
accuracies = [
    accuracy_score(y_test, baseline_pred),
    accuracy_score(y_test, lr_pred),
    accuracy_score(y_test, rf_pred),
    accuracy_score(y_test, xgb_pred)
]

# Plot
plt.figure(figsize=(8,5))
colors = ['gray', 'steelblue', 'green', 'orange']
bars = plt.bar(models, accuracies, color=colors)
plt.axhline(y=accuracies[0], color='red', linestyle='--', label='Naive Baseline')
plt.title('Model Accuracy Comparison — Nifty 50 Return Direction')
plt.ylabel('Accuracy')
plt.ylim(0.45, 0.60)
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002, 
             f'{acc:.3f}', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('../outputs/figures/accuracy_comparison.png', dpi=150)
plt.show()
print("Saved!")
```

## Results Interpretation
Logistic Regression achieved 53.9% accuracy, marginally beating the naive baseline of 52.8%.
Ensemble methods (Random Forest, XGBoost) did not outperform the baseline, suggesting
that market returns contain limited learnable signal from technical features alone.
This is consistent with the Efficient Market Hypothesis.

```python
from sklearn.metrics import RocCurveDisplay

fig, ax = plt.subplots(figsize=(8,6))

# Plot ROC curves for each model
RocCurveDisplay.from_estimator(lr, X_test_scaled, y_test, ax=ax, name='Logistic Regression')
RocCurveDisplay.from_estimator(rf, X_test, y_test, ax=ax, name='Random Forest')
RocCurveDisplay.from_estimator(xgb, X_test, y_test, ax=ax, name='XGBoost')

# Diagonal baseline
ax.plot([0,1], [0,1], 'k--', label='Random Classifier')
ax.set_title('ROC Curve Comparison — Nifty 50 Return Direction')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig('../outputs/figures/roc_curves.png', dpi=150)
plt.show()
print("Saved!")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, lr_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Down', 'Predicted Up'],
            yticklabels=['Actual Down', 'Actual Up'])
plt.title('Confusion Matrix — Logistic Regression')
plt.tight_layout()
plt.savefig('../outputs/figures/confusion_matrix.png', dpi=150)
plt.show()
print("Saved!")
```

```python
importances = rf.feature_importances_
feat_df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
}).sort_values('Importance', ascending=True)

plt.figure(figsize=(8,6))
plt.barh(feat_df['Feature'], feat_df['Importance'], color='steelblue')
plt.title('Feature Importance — Random Forest')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.savefig('../outputs/figures/feature_importance.png', dpi=150)
plt.show()
print("Saved!")
```

```python
results = pd.DataFrame({
    'Model': ['Naive Baseline', 'Logistic Regression', 'Random Forest', 'XGBoost'],
    'Accuracy': [0.528, 0.539, 0.506, 0.517],
    'ROC_AUC': ['-', '0.517', '0.500', '0.513']
})

import os
os.makedirs('../outputs/results', exist_ok=True)
results.to_csv('../outputs/results/model_comparison.csv', index=False)
print(results.to_string(index=False))
print("\nSaved!")
```
