# 01 — Data Collection
## Nifty 50 Return Direction Forecasting
This notebook downloads 11 years of Nifty 50 historical data (2015–2026) 
from Yahoo Finance and saves it for further processing.

**Data Source:** Yahoo Finance (^NSEI)  
**Period:** January 2015 — February 2026  
**Rows:** 2747 trading days

```python
import yfinance as yf
import pandas as pd

# Download Nifty 50 data
df = yf.download("^NSEI", start="2015-01-01", end="2026-02-27")
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
df.dropna(inplace=True)

print(df.head())
print(df.tail())
print(df.shape)
```

```python
df.to_csv('../data/nifty50_raw.csv')
print("Data saved successfully!")
```
